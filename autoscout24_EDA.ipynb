```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
import squarify as sq
```

```python
df = pd.read_csv("/content/drive/MyDrive/Data/autoscout24-cars.csv")
df
```

```python
df.info()
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df
```

```python
data = df.copy()
data
```

```python
# check for null values
data.isnull().sum()
```

```python
sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# drop null values
data = data.dropna()
data.isnull().sum()
print()
display(data)
```

```python
data.info()
```

```python
# drop duplicate rows
data = data.drop_duplicates().reset_index()
data
```

```python
data.drop(['index'], axis=1, inplace = True)
data
```

```python
# remove "€" from the front of price
data['price'] = [ab.split(" ")[1] for ab in data['price']]
```

```python
data
```

```python
data['power_pk'] = [pk.split('(')[-1].split(' ')[0] for pk in data['power']]
```

```python
data['power_kw'] = [kw.split(' ')[0] for kw in data['power']]
```

```python
data.drop(['power'], axis=1, inplace=True)
```

```python
data['kilometers'] = [km.split(' ')[0] for km in data['kilometers']]
```

```python
data
```

```python
data.info()
```

```python
# Changing Datatypes of some columns
#Price, Kilometers, Manufactured date
```

```python
data['manufactured_date'] = pd.to_datetime(data['manufactured_date'], format= '%m/%Y')
data
```

```python
data["year_manufactured"] = pd.DatetimeIndex(data['manufactured_date']).year
data["month_manufactured"] = pd.DatetimeIndex(data['manufactured_date']).month
data
```

```python
data.drop(['manufactured_date'], axis=1, inplace=True)
```

```python
price = []
for pr in data['price']:
  pr = pr.replace('.', '')
  price.append(pr)

data['price'] = price
data

kilom = []
for kilo in data['kilometers']:
  if '.' in kilo:
    k = kilo.replace('.', '')
  else:
    k = kilo
  kilom.append(k)

data['kilometers'] = kilom
data
```

```python
data['power_pk'] = data['power_pk'].astype('int')
data['power_kw'] = data['power_kw'].astype('int')
data['price'] = data['price'].astype('int')
data['kilometers'] = data['kilometers'].astype('int')
```

```python
data.tail()
```

```python
transmission = []
for transm in data['transmission']:
  if transm == 'Automatisch':
    transmission.append('Automatic')
  else:
    transmission.append('Manual')

data['transmission'] = transmission
data
```

```python
data['fuel_type'].value_counts()
```

```python
fuel_type = []
for f_type in data['fuel_type']:
  if f_type == 'Benzine':
    fuel_type.append('Petrol')
  elif f_type == 'Diesel':
    fuel_type.append('Diesel')
  elif f_type == 'Elektrisch/Benzine':
    fuel_type.append('Electric/Petrol')
  elif f_type == 'Elektrisch':
    fuel_type.append('Electric')
  elif f_type == 'Elektrisch/Diesel':
    fuel_type.append('Electric/Diesel')
  else:
    fuel_type.append('LPG')

data['fuel_type'] = fuel_type
data
```

```python
seller = []
for sell in data['seller']:
  if sell == 'Autobedrijf':
    seller.append('Car Company')

data['seller'] = seller
data
```

```python
data['name'] = data['name'].str.strip()
```

```python
# Top Most Expensive cars
expensive = data[['name_model', 'price']].sort_values(by=['price'], ascending=False).reset_index()
expensive.drop(['index'], axis=1, inplace=True)
expensive = expensive.head(20)
expensive
```

```python
plt.rcParams['figure.figsize'] = (15, 8)
x = list(expensive.price)
y = list(expensive.name_model)
plt.barh(y, x)
 
# setting label for each axis
plt.ylabel("Auto")
plt.xlabel("Price")
plt.title("Top Most Expensive Cars")
plt.show()
```

```python
# Cheapest Cars
cheap = data[['name_model', 'price']].sort_values(by=['price'], ascending=True).reset_index()
cheap.drop(['index'], axis=1, inplace=True)
cheap = cheap.head(10)
cheap = cheap.rename(columns={"name_model": "Auto", "price": "Price"})
cheap
```

```python
plt.rcParams['figure.figsize'] = (15, 8)
sns.barplot(x="Price", y='Auto', data=cheap, color="grey").set(title='Cheapest Cars')
plt.show()
```

```python
# Percentage of Auto by fuel type

fueltype_perc = data["fuel_type"].value_counts().reset_index()
fueltype_perc = fueltype_perc.rename(columns = {"index": "fuel_type", "fuel_type": "cars_count"})
fueltype_perc
```

```python
# Doughnut Chart of Fuel Types
colors = ["#003f5c", "#444e86", "#955196", "#dd5182", "#ff6e54", "#346888"]

# explosion
explode = (0.05, 0.05, 0.05, 0.05, 0.05, 0.05)
  
# Pie Chart
_, _, autotexts = plt.pie(fueltype_perc['cars_count'], colors=colors, labels=fueltype_perc['fuel_type'],
        autopct='%1.1f%%', pctdistance=0.85,
        explode=explode)
for ins in autotexts:
  ins.set_color('white')
  
# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Percentage Of Fuel Types')
plt.show()
```

```python
# Average price of brand/car by fuel type
avg_price = data.groupby(['fuel_type'])['price'].mean().reset_index()
avg_price
```

```python
fig = plt.figure(figsize = (15, 8))
 
# creating the bar plot
plt.bar(avg_price['fuel_type'], avg_price['price'], color ='maroon',
        width = 0.4)
 
plt.xlabel("Fuel Types")
plt.ylabel("Average price")
plt.title("Average Price Per Fuel Type")
plt.show()
```

```python
#Scatter plot to show correlation between Power and Price

data.plot.scatter(x = 'price', y = 'power_pk', s = 50, c = 'red')
plt.xlabel("Auto Price")
plt.ylabel("Horsepower")
plt.show()
```

```python
# Top 4 Cars with High Power/Horsepower
high_power = data.sort_values(by=['power_pk'], ascending=False).reset_index()
high_power.drop(columns=["index"], axis=1, inplace=True)
high_power = high_power.head(4)
high_power
```

```python
fig, axes = plt.subplots()

# using plt.stem
axes.stem(high_power['name_model'], high_power['power_pk'],
          use_line_collection=True, basefmt=' ')
  
# starting value of y-axis
axes.set_ylim(0)
  
# details and formatting of chart
plt.title('Top 4 Auto with High Power')
plt.xlabel('Auto')
plt.ylabel('Horsepower')
plt.xticks(high_power['name_model'], rotation = 45, fontsize= 13)
plt.show()
```

```python
# visualize amount of cars per kilometers bins
#create the bins
data['kilometers_bins'] = pd.cut(data['kilometers'], bins=6, labels=('L1', 'L2', 'L3', 'L4', 'L5', 'L6'))
```

```python
# get the count of each bin and visualie using a treemap
km_auto = data.groupby(['kilometers_bins'])['name_model'].count().reset_index()
km_auto = km_auto.rename(columns={'name_model': 'auto_count'})
km_auto

sq.plot(sizes=km_auto['auto_count'], label=km_auto['kilometers_bins'], alpha=.8 )
plt.axis('off')
plt.show()
```
